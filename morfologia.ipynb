```python
#No início do código é importado a biblioteca "cv2" para o processamento de imagem e o "numpy" para que seja feita a manipulação das matrizes.
import cv2
import numpy as np
#O "cv2_imshow" será utilizado para que as imagem sejam exibidas no Google Colab.
from google.colab.patches import cv2_imshow

#Nesse trecho é lidoas imagens com o "cv2.imread()"sendo atribuida a alguma variável.
#O "j" fica atribuido a "img", "j_ruido" à "img_opening" e "j_furos" à "img_closing".
#O número "0" (zero) é um parametro a qual indica que a imagem deve ser carregada em modo de escala de cinza.
img = cv2.imread('j.png', 0)
img_opening = cv2.imread('j_ruido.png', 0)
img_closing = cv2.imread('j_furos.png', 0)

#Nessa parte do código é obtido as dimenssões das "img" usando a variavel "shape".
altura, largura = img.shape

#Aqui cria-se uma matris 5x5 preenchida com uns (1s). O np.ones é uma função da biblioteca NumPy que retorna uma matriz preenchida com um determinado valor.
#O "np.uint8" é o tipo de dado utilizado e são números inteiros sem sinal de 8 bits.
kernel = np.ones((5,5),np.uint8)
#Aqui imprime o conteúdo da matriz kernel
print(kernel)

#Aqui é Realizado a erosão e dilatação de das imagens, e será atribuidas às variáveis "erosion" e "dilation"
#A erosão e a dilatação estão sendo aplicadas duas vezes indicado pelo "iterations = 2"
erosion = cv2.erode(img,kernel,iterations = 2)
dilation = cv2.dilate(img,kernel,iterations = 2)

#Aqui é aplicada a operação de gradiente morfológico na imagem "img"
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
#Nessa parte é aplicada a operação de abertura morfológica na imagem removendo ruídos e objetos pequenos
opening = cv2.morphologyEx(img_opening, cv2.MORPH_OPEN, kernel)
#Aqui é a parte final a qual faz o fechamento que nada mais é que preencher lacunas em objetos e conectar componentes quebrados da imagem
closing = cv2.morphologyEx(img_closing, cv2.MORPH_CLOSE, kernel)

#Impressão da imagem separadas
cv2_imshow(img)
cv2_imshow(erosion)
cv2_imshow(dilation)
cv2_imshow(opening)
cv2_imshow(closing)
cv2_imshow(gradient)
```
