```python
#No incio do cdigo  importado a biblioteca "cv2" para o processamento de imagem e o "numpy" para que seja feita a manipulao das matrizes.
import cv2
import numpy as np
#O "cv2_imshow" ser utilizado para que as imagem sejam exibidas no Google Colab.
from google.colab.patches import cv2_imshow

#Nesse trecho  lidoas imagens com o "cv2.imread()"sendo atribuida a alguma varivel.
#O "j" fica atribuido a "img", "j_ruido"  "img_opening" e "j_furos"  "img_closing".
#O nmero "0" (zero)  um parametro a qual indica que a imagem deve ser carregada em modo de escala de cinza.
img = cv2.imread('j.png', 0)
img_opening = cv2.imread('j_ruido.png', 0)
img_closing = cv2.imread('j_furos.png', 0)

#Nessa parte do cdigo  obtido as dimensses das "img" usando a variavel "shape".
altura, largura = img.shape

#Aqui cria-se uma matris 5x5 preenchida com uns (1s). O np.ones  uma funo da biblioteca NumPy que retorna uma matriz preenchida com um determinado valor.
#O "np.uint8"  o tipo de dado utilizado e so nmeros inteiros sem sinal de 8 bits.
kernel = np.ones((5,5),np.uint8)
#Aqui imprime o contedo da matriz kernel
print(kernel)

#Aqui  Realizado a eroso e dilatao de das imagens, e ser atribuidas s variveis "erosion" e "dilation"
#A eroso e a dilatao esto sendo aplicadas duas vezes indicado pelo "iterations = 2"
erosion = cv2.erode(img,kernel,iterations = 2)
dilation = cv2.dilate(img,kernel,iterations = 2)

#Aqui  aplicada a operao de gradiente morfolgico na imagem "img"
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
#Nessa parte  aplicada a operao de abertura morfolgica na imagem removendo rudos e objetos pequenos
opening = cv2.morphologyEx(img_opening, cv2.MORPH_OPEN, kernel)
#Aqui  a parte final a qual faz o fechamento que nada mais  que preencher lacunas em objetos e conectar componentes quebrados da imagem
closing = cv2.morphologyEx(img_closing, cv2.MORPH_CLOSE, kernel)

#Impresso da imagem separadas
cv2_imshow(img)
cv2_imshow(erosion)
cv2_imshow(dilation)
cv2_imshow(opening)
cv2_imshow(closing)
cv2_imshow(gradient)
```
